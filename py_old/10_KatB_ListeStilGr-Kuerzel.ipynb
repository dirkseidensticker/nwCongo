# Liste der Stilgruppen-Kürzel

```python
%matplotlib inline
from IPython.display import display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3

con = sqlite3.connect("../data/CongoDB.sqlite")
```

```python
df = pd.read_csv("../lit/Wotzka1995_StilGrKuerzel.csv")

df["Wotzka1995"] = df["Wotzka1995"].replace(np.nan, "-")
df["Seidensticker"] = df["Seidensticker"].replace(np.nan, "-")
df["Lit"] = df["Lit"].replace(np.nan, "-")

df["Lit"] = df["Lit"].replace('EggertKanimba1980', "\cite{Eggert.1980c}")

df = df.rename(columns={'Typ': 'Kürzel',
                        'Wotzka1995': 'Wotzka 1995',
                        'Seidensticker': 'Seidensticker',
                        'Lit': 'Andere'})

df.to_latex('../output/tabs/10_KatB_ListeStilGr-Kuerzel.tex', 
            longtable = True, index = False, escape = False, encoding = 'utf8')

df
```
