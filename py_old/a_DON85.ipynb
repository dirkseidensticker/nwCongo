```python
%matplotlib inline
from IPython.display import display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3

con = sqlite3.connect("../data/CongoDB.sqlite")
```

## Stilgruppen

```python
sql = """SELECT
           t_Obj.objID,
           t_Obj.Art,
           t_Komplex.bef_nr,
           t_Obj.Typ
       FROM (t_Ort INNER JOIN t_Komplex ON t_Ort.ortID = t_Komplex.ortID)
           INNER JOIN t_Obj ON t_Komplex.komplexID = t_Obj.komplexID
       WHERE (((t_Ort.ort_kurz) == 'DON')
           AND ((t_Obj.Art) == 'K'))""".replace('\n',' ')

df = pd.read_sql(sql, con)

df_pivot = df.pivot_table(values = 'objID', index = 'Typ', columns = 'bef_nr', aggfunc = len)
```

```python
fig = plt.figure(1)
fig.set_size_inches(12, 5.5)

plt.subplot(121, aspect='equal')
df_pivot['85/101'].plot(kind='pie', autopct='%.0f %%', cmap = 'Paired')
plt.ylabel('')
plt.title('DON 85/101')

plt.subplot(122, aspect='equal')
df_pivot['85/102'].plot(kind='pie', autopct='%.0f %%', cmap = 'Paired')
plt.ylabel('')
plt.title('DON 85/102')
```
